EXE-01:SCATTERPLOT-BASIC

```python
people = ['Ammu','jaanu','spandana','manasa']
age = [21,12,23,12]
weight = [55,45,65,50]
height = [160,155,140,139]


import matplotlib.pyplot as plt
plt.scatter(weight,height)
plt.scatter(age,weight)
plt.scatter(age,height)
plt.show
```

```python
plt.figure(figsize=(6,6))
plt.title("people's weight in kgs",fontsize=15,fontstyle="italic")
plt.bar(x=people,height=weight)

plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=45)
plt.yticks(fontsize=14)
plt.ylabel("weight in kgs",fontsize=15)
plt.show
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color="red",edgecolor="k",alpha=0.75,bins=5)
plt.title("histogram of patient weight",fontsize=18)
plt.xlabel("weight (in kgs)",fontsize=18)
plt.xticks(fontsize=10,color='blue')
plt.show
```
