## Activity-02 Data visualisation

**Scatter plot


```python
people = ['pawan','Kalyan','Mahi','Dhoni']
age = [21,12,3,22]
weight = [43,35,61,64]
height = [135,145,120,110]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,4))
plt.title("People's weight in kgs vs height in cms")
plt.xlabel("weight")
plt.ylabel("height")
plt.scatter(weight, height)
plt.show()
```

**Bar chart

```python
plt.figure(figsize=(4,4))
plt.title("People's weight in kgs")
plt.bar(x=people,height = weight)
plt.xlabel("People")
plt.ylabel("Weight")
```

**Histogram

```python
import numpy as np
plt.figure(figsize = (7,5))
plt.hist(weight,edgecolor = 'black' ,bins = 5)
plt.title("Histogram of patient weight")
plt.xlabel("Weight in kgs")
plt.xticks()
plt.yticks()
plt.show
```

```python
import numpy as np
plt.figure(figsize = (7,5))
plt.hist(age,edgecolor = 'black' ,bins = 5)
plt.title("Histogram of patient age")
plt.xlabel("Age")
plt.xticks()
plt.yticks()
plt.show
```

```python
import numpy as np
plt.figure(figsize = (7,5))
plt.hist(height,edgecolor = 'black' ,bins = 5)
plt.title("Histogram of patient height")
plt.xlabel("Height in kgs")
plt.xticks()
plt.yticks()
plt.show
```
